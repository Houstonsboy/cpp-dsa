```c++
#include <iostream>
using namespace std;

    int x = 0;
    int y = 20;
    int k = 10; 
    while (k > 0) {
        y = y - 4;
        
        //prevents division by zero
        if (y == 0) {
            break; 
        }
        x = (2 / y) + x;
        if (y < 6) {
            break; // break loop is condition iis met
        }
        k--; // Decrement i so the loop eventually ends
    }

    cout << "Final x: " << x << endl;
    return 0;
```

```c++
#include <iostream>
using namespace std;

int q = 0;
int sum = 0;
int myArray[5];
//loop 5 times getting input and adding it to sum to get total
while (q < 5) {
    cout << "please input array index " << q << ": ";
    cin >> myArray[q];
    sum = sum + myArray[q];
    q++;
}

double average = (double)sum / 5;

cout << "the total is: " << sum << endl;
cout << "the average is: " << average << endl;
```
